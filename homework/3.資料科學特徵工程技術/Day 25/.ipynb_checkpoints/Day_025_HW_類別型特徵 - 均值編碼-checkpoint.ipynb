# 作業1
* 請仿照範例，將鐵達尼範例中的類別型特徵改用均值編碼實作一次

```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
import copy, time
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder

data_path = 'C:/Users/User/Desktop/Github/ML100Days/homework/'
df_train = pd.read_csv(data_path + 'titanic_train.csv')
df_test = pd.read_csv(data_path + 'titanic_test.csv')

train_Y = df_train['Survived']
ids = df_test['PassengerId']
df_train = df_train.drop(['PassengerId', 'Survived'] , axis=1)
df_test = df_test.drop(['PassengerId'] , axis=1)
df = pd.concat([df_train,df_test])   #合併處理，做特徵工程
df.head()
```

```python
#只取類別值 (object) 型欄位, 存於 object_features 中
object_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'object':
        object_features.append(feature)
print(f'{len(object_features)} Numeric Features : {object_features}\n')

# 只留類別型欄位
df = df[object_features]
df = df.fillna('None')
train_num = train_Y.shape[0]
df.head()
```

```python
df.info()
```

# 作業2
* 觀察鐵達尼生存預測中，均值編碼與標籤編碼兩者比較，哪一個效果比較好? 可能的原因是什麼?

```python
# 對照組 : 標籤編碼 + 邏輯斯迴歸

df_temp = pd.DataFrame()
print(df_temp)
print("==============================================================")

for c in df.columns:
    df_temp[c] = LabelEncoder().fit_transform(df[c])

print(df_temp)
```

```python
train_X = df_temp[:train_num] #拆分出訓練資料集
train_X
```

```python
estimator = LogisticRegression()
start = time.time()
print(f'shape : {train_X.shape}')
print(f'score : {cross_val_score(estimator, train_X, train_Y, cv=5).mean()}')
print(f'time : {time.time() - start} sec')
```

```python
# 均值編碼 + 邏輯斯迴歸

data = pd.concat([df[:train_num], train_Y], axis=1)
data
```

```python
for c in df.columns:
    mean_df = data.groupby([c])['Survived'].mean().reset_index() 
    #根據每一個欄位，去對目標變數欄位做groupby算平均
    #5個變數欄位 >> 5張樞紐分析表
    print(mean_df)
    mean_df.columns = [c, f'{c}_mean']  #5個新欄位的名稱都是原本名稱後面加上"_mean"
    data = pd.merge(data, mean_df, on=c, how='left') 
    #將5張樞紐分析表根據欄位，和原本的資料框data做合併
    #因為5章收紐分析表列數各異(小於等於data的列數)，故使用左合併:
    # https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.merge.html
    data = data.drop([c] , axis=1) #合併完成後就可以丟掉原本欄位了
    
print("==========================================================")    
print(data)
```

```python
data = data.drop(['Survived'] , axis=1) #把目標函數也丟掉，才算是完成了訓練資料集
```

```python
estimator = LogisticRegression()
start = time.time()
print(f'shape : {train_X.shape}')
print(f'score : {cross_val_score(estimator, data, train_Y, cv=5).mean()}')
print(f'time : {time.time() - start} sec')
```

透過均值編碼之後，模型表現直接從0.77變成了1.0!

這主要是因為'Name'與'Ticket'兩個欄位的種類數眾多
如果均值編碼時加入這裡個欄位的均值會過擬合( overfitting )
因此下列程式碼中，需要先移除這兩個欄位，再做均值編碼

```python
# 對照組 : 標籤編碼 + 邏輯斯迴歸
df_temp = pd.DataFrame()
for c in df.columns:
    df_temp[c] = LabelEncoder().fit_transform(df[c])
train_X = df_temp[:train_num]
estimator = LogisticRegression()
start = time.time()
print(f'shape : {train_X.shape}')
print(f'score : {cross_val_score(estimator, train_X, train_Y, cv=5).mean()}')
print(f'time : {time.time() - start} sec')
```

```python
# 如果欄位類別數太多, 需要先剔除, 否則會 overfit (正確率會到達 1.0)
train_X.nunique()
```

```python
# 均值編碼 + 邏輯斯迴歸

data = pd.concat([df[:train_num], train_Y], axis=1)

for c in df.columns:
    mean_df = data.groupby([c])['Survived'].mean().reset_index()
    mean_df.columns = [c, f'{c}_mean']
    data = pd.merge(data, mean_df, on=c, how='left')
    data = data.drop([c] , axis=1)
    
data = data.drop(['Survived', 'Name_mean', 'Ticket_mean'] , axis=1)

estimator = LogisticRegression()
start = time.time()
print(f'shape : {train_X.shape}')
print(f'score : {cross_val_score(estimator, data, train_Y, cv=5).mean()}')
print(f'time : {time.time() - start} sec')
```

從0.77進步到0.83

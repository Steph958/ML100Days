# 讀取資料 & 載入套件:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

train = pd.read_csv('C:/Users/USER/Desktop/Github/ML100Days/train.csv',header=None)
test = pd.read_csv('C:/Users/USER/Desktop/Github/ML100Days/test.csv',header=None)
trainLabel = pd.read_csv('C:/Users/USER/Desktop/Github/ML100Days/trainLabels.csv',header=None)

print('train shape:', train.shape)
print('test shape:', test.shape)
print('trainLabel shape:', trainLabel.shape)
train.head()
```

# 建立一個Baseline:

```python
from sklearn.model_selection import cross_val_score, train_test_split

X, y = train, np.ravel(trainLabel)
print(X.shape)
print("=====================================================")
print(y)

print("=====================================================")
print(trainLabel)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, trainLabel, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

clf = RandomForestClassifier()

clf.fit(X_train, y_train)

print(clf)

pred1 = clf.predict(X_test)

acc_baseline = metrics.accuracy_score(y_test, pred1)
print("Baseline Accuracy: ", acc_baseline)
```

```python
#視覺化:

tree_num = np.arange(10,100)

train_accuracy =[]


#Loop
for i,k in enumerate(tree_num):
    clf = RandomForestClassifier(n_estimators=k)
    clf.fit(X_train,y_train)
    pred = clf.predict(X_test)
    acc = metrics.accuracy_score(y_test,pred)
    train_accuracy.append(acc)
    
    
#Plot
plt.figure(figsize=[12,8])
plt.plot(tree_num, train_accuracy, label='Training Accuracy')
plt.legend()
plt.title('Random Forest Model:Number of trees VS Accuracy')
plt.xlabel('Number of trees')
#plt.xticks(tree_num)
plt.ylabel('Accuracy')
plt.show()

```

```python
clf = RandomForestClassifier(n_estimators=98)

clf.fit(X_train, y_train)

print(clf)

pred2 = clf.predict(X_test)

acc_1 = metrics.accuracy_score(y_test, pred2)
print("Tuned Accuracy: ", acc_1)
```

```python
actual_pred = clf.predict(test)
len(actual_pred)
```

```python
ID = range(1,9001)
submit = pd.DataFrame(ID, columns = ['ID'],index = None)
submit['Solution'] = actual_pred
submit
```

使用function y=(x+5)² 來練習學習率( leanrning rate ) 調整所造成的影響

# 作業目標:
    請嘗試使用不同的組合驗證 learning rate 對所需 iteration 的影響
    lr = [0.1, 0.0001]
    主要驗證 Lr 對於grandient 收斂的速度
    

```python
cur_x = 3 # The algorithm starts at x=3

precision = 0.000001 #This tells us when to stop the algorithm

max_iters = 10000 # maximum number of iterations

df = lambda x: 2*(x+5) #Gradient of our function # dy/dx #匿名函數
```

```python
#繪圖用:
iters_history = []
x_history = []
```

```python
#需要執行初始值設定, 下列三個:
# Learning rate
lr = 0.03

#iteration counter
iters = 0  

#Step size  
previous_step_size = 1
```

如果step size還未降到標準(precision)以下、且迭代次數也未超過設定(max_iters)的話，進行迴圈內容執行:

```python

while previous_step_size > precision and iters < max_iters:
    
    #進行梯度下降
    
    prev_x = cur_x    #要帶入的x值 #起始值設定為3
    
    cur_x = cur_x - lr*df(prev_x) # Gradient Descent
    
    previous_step_size = abs(cur_x - prev_x) #GD後減GD前，更新步伐
    
    iters+=1
    
    print("the learning rate is",lr, "\nIteration",iters,"\nX value is",cur_x) 
    
    #繪圖用的資訊:
    iters_history.append(iters)
    x_history.append(cur_x)
```

```python
print("Totally iteations: ", iters)
print("The local minimum occurs at", cur_x)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(iters_history, x_history, 'o-', ms=2, lw=1.0, color='b')
plt.xlabel(r'$iters$', fontsize=15)
plt.ylabel(r'$x$', fontsize=15)
plt.show()
```

lr=0.01 >> iterations=595

lr=0.03 >> iterations=213

學習率設定較高的話，可以較快的收斂結果(所需的迭代次數較少)
